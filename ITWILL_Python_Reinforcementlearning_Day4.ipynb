## ■ ping pong 게임을 하려면 필요한것이 무엇인가?

1. 켄버스(canvas)
2. 공(ball) 클래스
    - init 함수
    - 공을 움직이게 하는 함수
    - 패들에 부딪히면 공이 튀기게 하는 함수


3. 패들(paddle) 클래스
    - init 함수(패들의 색깔과 크기, 캔버스에서의 위치
    - 패들이 움직이게 하는 함수
        1. 패들이 캔버스 밖으로 안나가게 하는 코드
    - 왼쪽으로 움직이게 하는 함수
    - 오른쪽으로 움직이게 하는 함수
    

## ■ ping pong게임을 머신러닝화 하는 방법

1. ping pong 게임만 되게하는 클래스
    - canvas 클래스 : 캔버스의 크기, 캔버스의 색깔
    - ball 클래스 
        - init 함수 : 공 색깔, 공 크기, 공 위치
        - draw 함수 : 공이 화면 밖으로 안나가게하는 함수
        - hit_paddle 함수 : 공이 패들에 닿으면 튀어오르게 하는 함수
    - paddle 클래스
        - init 함수 : 패들의 크기, 위치, 키보드 바인딩
        - draw 함수 : 패들이 화면 밖으로 안나가게 하는 함수
        - right_shift : 패들을 오른쪽으로 움직이게 하는 함수
        - left_shift : 패들을 왼쪽으로 움직이게 하는 함수
        
        
2. ping pong 게임을 학습시키는 함수(MDP의 구성요소들을 가지고 학습데이터를 만드는 함수들)
    1. keystat 함수 : 학습 데이터의 모양을 만드는 함수
    2. add 함수 : 학습 데이터를 values 딕셔너리에 추가하는 함수
    3. lookup 함수 : 학습 데이터를 values 딕셔너리에서 찾아보고 못찾았으면 새로 add 하는 함수
    4. greedy 함수 : 현재 상황에서 가장 좋은 방향을 찾는 함수
    5. backup 함수 : 학습 데이터의 가치 함수의 수학식이 구현되어 있는 함수
    6. action 함수 : Random move와 Greedy move중에 하나를 결정하는 함수 
    7. winnerval 함수 : 가중치를 보상하는 함수(1, -1, 0)
    8. cyclestate 함수 : 패들의 위치부터 학습 데이터를 모으겠금 해주는 함수
    
    
3. 부가적인 함수
    1. gameover 함수 : 볼이 한 사이클을 돌아서 hit 했는지 miss 했는지를 알려주는 함수
    2. randomChoice 함수 : 0 또는 1을 랜덤으로 선택하게 하는 함수
    3. writecsv 함수 : 학습 데이터를 csv로 저장시키는 함수
    4. loadcsv 함수 : 학습 데이터를 values 딕셔너리에 읽어들이는 함수

## 문제1. 캔버스를 그리시오!

```python
from tkinter import * 
import random
import time


tk = Tk()      # 1. tk 를 인스턴스화 한다.

tk.title("Ping Pong Game")  # 2. tk 객체의 title 메소드(함수)로 게임창에 제목을 부여한다.

tk.resizable(0, 0) # 3. 게임창의 크기는 가로나 세로로 변경될수 없다라고 말하는것이다. 

tk.wm_attributes("-topmost", 1) # 4. 다른 모든 창들 앞에 캔버스를 가진 창이 위치할것을 tkinter 에게 알려준다. 

canvas = Canvas(tk, width=500, height=400, bd=0, highlightthickness=0) 

# bd=0,highlightthickness=0 은 캔버스 외곽에 둘러싼 

# 외곽선이 없도록 하는것이다. (게임화면이 좀더 좋게)



canvas.configure(background='black')  # 캔버스 색깔 변경하는 방법 

canvas.pack()       # 앞의 코드에서 전달된 폭과 높이는 매개변수에 따라 크기를 맞추라고 캔버스에에 말해준다.                                                                     

tk.update()   # tkinter 에게 게임에서의 애니메이션을 위해 자신을 초기화하라고 알려주는것이다.
```

## 문제6. 공을 움직이게 하시오!

```python
from tkinter import * 
import random
import time


class Ball:
    
    def __init__(self, canvas, color):
        self.canvas = canvas
        self.id = canvas.create_oval(10, 10, 25, 25, fill=color) #공 좌표 및 색깔(oval : object 형태 타입)
                               #        ↑   ↑  ↑    ↑
      #좌측상단  구석에서 시작하는  x  y   우측 하단 구석에서 시작하는 x y
        self.canvas.move(self.id, 245, 100) #공을 캔버스 중앙으로 이동

    def draw(self):
        self.canvas.move(self.id, 0, -1) 
                      # 공 자신, 가로, 세로
                      # 0 은 가로로 움직이지 말아라 !
                      # -1 은 화면의 위쪽으로 1픽셀씩 움직여라

tk = Tk()      # 1. tk 를 인스턴스화 한다.
tk.title("Ping Pong Game")  # 2. tk 객체의 title 메소드(함수)로 게임창에 제목을 부여한다.
tk.resizable(0, 0) # 3. 게임창의 크기는 가로나 세로로 변경될수 없다라고 말하는것이다. 
tk.wm_attributes("-topmost", 1) #4. 다른 모든 창들 앞에 캔버스를 가진 창이 위치할것을 tkinter 에게 알려준다. 
canvas = Canvas(tk, width=500, height=400, bd=0, highlightthickness=0) 
# bg=0,highlightthickness=0 은 캔버스 외곽에 둘러싼 
# 외곽선이 없도록 하는것이다. (게임화면이 좀더 좋게)


canvas.configure(background='black')  # 캔버스 색깔 변경하는 방법 
canvas.pack()       # 앞의 코드에서 전달된 폭과 높이는 매개변수에 따라 크기를 맞추라고 캔버스에에 말해준다.                                                                     
tk.update()   # tkinter 에게 게임에서의 애니메이션을 위해 자신을 초기화하라고 알려주는것이다.   

         
ball = Ball(canvas,'blue')

while 1:
    ball.draw()
    tk.update_idletasks()   # 우리가 창을 닫으라고 할때까지 계속해서 tkinter 에게 화면을 그리고 
    tk.update() 
    time.sleep(0.01)  # 100분의 1초마다 잠들어라 ! 
```

## 문제7. 게임시작할때 공이 대각선 위의 방향으로 올라가게 하시오!

```python
from tkinter import * 
import random
import time


class Ball:
    
    def __init__(self, canvas, color):
        self.canvas = canvas
        self.id = canvas.create_oval(10, 10, 25, 25, fill=color) #공 좌표 및 색깔(oval : object 형태 타입)
                               #        ↑   ↑  ↑    ↑
      #좌측상단  구석에서 시작하는  x  y   우측 하단 구석에서 시작하는 x y
        self.canvas.move(self.id, 245, 100) #공을 캔버스 중앙으로 이동

    def draw(self):
        self.canvas.move(self.id, 1, -1) 
                      # 공 자신, 가로, 세로
                      # 0 은 가로로 움직이지 말아라 !
                      # -1 은 화면의 위쪽으로 1픽셀씩 움직여라

tk = Tk()      # 1. tk 를 인스턴스화 한다.
tk.title("Ping Pong Game")  # 2. tk 객체의 title 메소드(함수)로 게임창에 제목을 부여한다.
tk.resizable(0, 0) # 3. 게임창의 크기는 가로나 세로로 변경될수 없다라고 말하는것이다. 
tk.wm_attributes("-topmost", 1) #4. 다른 모든 창들 앞에 캔버스를 가진 창이 위치할것을 tkinter 에게 알려준다. 
canvas = Canvas(tk, width=500, height=400, bd=0, highlightthickness=0) 
# bg=0,highlightthickness=0 은 캔버스 외곽에 둘러싼 
# 외곽선이 없도록 하는것이다. (게임화면이 좀더 좋게)


canvas.configure(background='black')  # 캔버스 색깔 변경하는 방법 
canvas.pack()       # 앞의 코드에서 전달된 폭과 높이는 매개변수에 따라 크기를 맞추라고 캔버스에에 말해준다.                                                                     
tk.update()   # tkinter 에게 게임에서의 애니메이션을 위해 자신을 초기화하라고 알려주는것이다.   

         
ball = Ball(canvas,'blue')

while 1:
    ball.draw()
    tk.update_idletasks()   # 우리가 창을 닫으라고 할때까지 계속해서 tkinter 에게 화면을 그리고 
    tk.update() 
    time.sleep(0.01)  # 100분의 1초마다 잠들어라 ! 
```

```python
from tkinter import * 
import random
import time


class Ball:
    
    def __init__(self, canvas, color):
        self.canvas = canvas
        self.id = canvas.create_oval(10, 10, 25, 25, fill=color) #공 좌표 및 색깔(oval : object 형태 타입)
                               #        ↑   ↑  ↑    ↑
      #좌측상단  구석에서 시작하는  x  y   우측 하단 구석에서 시작하는 x y
    
        self.canvas.move(self.id, 245, 100) #공을 캔버스 중앙으로 이동
        self.x = 0  # 패들이 처음 시작할때 움직이지 않게 0으로 설정
        self.y = 0
        self.canvas.bind_all('<KeyPress-Left>', self.turn_left)  # 왼쪽 화살표 키를 '<KeyPress-Left>'  라는 이름로 바인딩
        self.canvas.bind_all('<KeyPress-Right>', self.turn_right)  # 오른쪽도 마찬가지로 바인딩한다.
        self.canvas.bind_all('<KeyPress-Up>', self.turn_up)  # 위쪽 화살표 키를 '<KeyPress-Up>'  라는 이름로 바인딩
        self.canvas.bind_all('<KeyPress-Down>', self.turn_down)  # 아래쪽도 마찬가지로 바인딩한다.
        
    def draw(self):
        self.canvas.move(self.id, self.x, self.y) 
    
    def turn_left(self, evt):  # 패들의 방향을 전환하는 함수
        self.x = -5

    def turn_right(self, evt):
        self.x = 5
        
    def turn_up(self, evt):  # 패들의 방향을 전환하는 함수
        self.y = -5

    def turn_down(self, evt):
        self.y = 5
        

tk = Tk()      # 1. tk 를 인스턴스화 한다.
tk.title("Ping Pong Game")  # 2. tk 객체의 title 메소드(함수)로 게임창에 제목을 부여한다.
tk.resizable(0, 0) # 3. 게임창의 크기는 가로나 세로로 변경될수 없다라고 말하는것이다. 
tk.wm_attributes("-topmost", 1) #4. 다른 모든 창들 앞에 캔버스를 가진 창이 위치할것을 tkinter 에게 알려준다. 
canvas = Canvas(tk, width=500, height=400, bd=0, highlightthickness=0) 
# bg=0,highlightthickness=0 은 캔버스 외곽에 둘러싼 
# 외곽선이 없도록 하는것이다. (게임화면이 좀더 좋게)


canvas.configure(background='white')  # 캔버스 색깔 변경하는 방법 
canvas.pack()       # 앞의 코드에서 전달된 폭과 높이는 매개변수에 따라 크기를 맞추라고 캔버스에에 말해준다.                                                                     
tk.update()   # tkinter 에게 게임에서의 애니메이션을 위해 자신을 초기화하라고 알려주는것이다.   

         
ball = Ball(canvas,'blue')

while 1:
    ball.draw()
    tk.update_idletasks()   # 우리가 창을 닫으라고 할때까지 계속해서 tkinter 에게 화면을 그리고 
    tk.update() 
    time.sleep(0.01)  # 100분의 1초마다 잠들어라 ! 
```
